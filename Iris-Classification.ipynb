```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
iris['species'] = le.fit_transform(iris['species'])
```

```python
iris
```

```python
iris['species'].value_counts()
```

# now model training

```python
from sklearn.model_selection import train_test_split
x = iris.drop('species', axis=1)
```

```python
y = iris['species']
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
models = { 
    'logistic':LogisticRegression(max_iter=200),   # showed limit of default 100 iterations
    'KNN': KNeighborsClassifier(),
    'descision': DecisionTreeClassifier()
}
```

```python
for name , m in models.items():
    m.fit(x_train,y_train)
    predict = m.predict(x_test)
    print(name , "accuracy", m.score(x_test,y_test))
```

```python
cv_results={}  # to store cross validation results
from sklearn.model_selection import cross_val_score  
```

```python
for name,m in models.items():
    scores = cross_val_score(m , x , y , cv=5 )  # 5 fold cross validation
    cv_results[name] = scores.mean()
    print(f"{name} = {scores.mean():4f}")
```

```python
plt.figure(figsize=(7,5))
plt.bar(cv_results.keys(), cv_results.values() , color=['skyblue','orange','green'])
plt.ylabel("Mean Accuracy")
plt.xlabel("Models")
plt.title("Model Comparison (5-Fold Cross-Validation)")
plt.ylim(0.8, 1)
plt.show()
```
